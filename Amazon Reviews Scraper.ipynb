```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import datetime
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \AppleWebKit/537.36 (KHTML, like Gecko) \Chrome/90.0.4430.212 Safari/537.36'}

product_links = []

for x in range(0,23):
    links = f'https://www.amazon.in/Apple-iPhone-13-Pro-256GB/product-reviews/B09G99BKK4/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={x}'
    product_links.append(links)
    
for links in product_links:
    r = requests.get(links)
    soup = BeautifulSoup(r.content)
    Review_Title = soup.find('div' , class_ = 'a-section a-spacing-none review-views celwidget')
    for title in Review_Title:
        Title = title.find('a' , class_ = 'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold')
        try:
            print(Title.text.strip())
        except:
            print(" ")
```
